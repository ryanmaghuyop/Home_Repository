### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


````python
#Initializing packages
import numpy as np
import matplotlib.pyplot as plt
from cmath import exp,pi,sin
from numpy.fft import rfft,irfft
````

For a given function $f(x)$, we can take its Fourier series expansion as

\begin{equation}
f(x) = \sum_{k = -\infty}^{\infty} \gamma_k \exp \left( i \frac{2 \pi k x}{L} \right)
\end{equation}

where $\gamma_k$ are the *Fourier coefficients* given by

\begin{equation}
\gamma_k = \frac{1}{L} \int_0^L f(x) \exp \left( - i \frac{2 \pi k x}{L} \right) dx
\end{equation}

Taking discrete points $x_n$ and defining $y_n := f(x_n)$ as samples, we can drop the factor 1/N to get the standard quantity in the discrete Fourier transform $c_k$.

\begin{equation}
c_k = \sum_{n=0}^{N-1} y_n \exp \left( - i \frac{2 \pi k n}{N} \right)
\end{equation}

These coefficients $c_k$ are complex-valued, with half of the coefficients being the complex conjugates of the other half. Thus, if we have $N$ number of discrete points, we end up having $\frac{N}{2} + 1 $ coefficients for even N or $\frac{(N+1)}{2}$ coefficients for odd N.

````python
N = 1000 #Given number of slices

#Defined functions
def func(n): #Given modulated sine wave
  f = sin(pi*n/N)*sin(20*pi*n/N)
  return f

def coeff(y): #Function that gives the j^{th} coefficient of the DFT for a given function y[n]
  c = np.zeros(N//2+1,complex)
  for j in range(N//2+1):
    for n in range(N):
      c[j] += y[n]*exp(-2j*pi*j*n/N)
  return c

#-----------------------------------------------------------
#Variables for plotting
n = np.arange(0, 1001, 1) #Array of values from 0 to N
y = [] #Array of values of the given sine function evaluated from 0 to N
for j in n:
  y.append(func(j))

kpoints = np.arange(0, 501, 1) #Array of values from 0 to N//2 + 1
c = coeff(y) #Array of the coefficients of the DFT

#-----------------------------------------------------------
#Plots

fig, ax = plt.subplots()
ax.plot(n, y, color = 'blue')
ax.set_xlabel("Time (s)")
ax.set_ylabel("$y$")
ax.set_title("Modulated sine wave signal")
plt.show()

fig1, ax1 = plt.subplots()
ax1.plot(kpoints, abs(c), color = 'green')
ax1.annotate("Maximum of " + str(round(np.max(abs(c)), 2)) + " at $\omega = 10$", xy=(10, np.max(abs(c))) , xytext=(15,280), arrowprops={"arrowstyle":"->", "color":"black"})
ax1.set_xlabel("$\omega$ $(\dfrac{1}{s})$")
ax1.set_ylabel("$|c_k|$")
ax1.set_title("Discrete Fourier transform of the sine wave signal")
plt.show()

fig2, ax2 = plt.subplots()
ax2.plot(abs(c), color = 'green')
plt.xlim(5,15)
ax2.annotate("Maximum of " + str(round(np.max(abs(c)), 2)) + " at $\omega = 10$", xy=(10, np.max(abs(c))) , xytext=(5.25,250), arrowprops={"arrowstyle":"->", "color":"black"})
ax2.set_xlabel("$\omega$ $(\dfrac{1}{s})$")
ax2.set_ylabel("$|c_k|$")
ax2.set_title("Discrete Fourier transform of the sine wave signal (zoomed in)")
plt.show()
````

As seen from the graph, almost all $c_k \approx 0$ for all $k$, except $k \in [8,12]$. Thus, we can approximate the given $y_n$ using the inverse DFT

\begin{equation}
  \begin{split}
    y_n &= \frac{1}{N} \sum_{k=0}^{N-1}c_k \exp \left( i \frac{2 \pi k n}{N} \right) \\
    &\sim \frac{1}{1000} \sum_{k = 8}^{12}c_k \exp \left( i \frac{\pi k n}{500} \right)
  \end{split}
\end{equation}

where $c_k$ can be obtained by calling *c[8:13]*.

<span style="color:blue"> Since the product of two sines is the sum of two pure sine frequencies, you can show that there are two pure frequencies at 9.5 Hz and 10.5 Hz. Since the DFT does not sample exactly at these frequencies (they are "in-between" the sample points), you observe this envelope around 10Hz.</span>

````python
ck = rfft(dow) #Full array of coefficients of the DFT for the data
ck_10pct = rfft(dow) #Array with only the first 10% of the data
ck_5pct = rfft(dow) #Array with only the first 5% of the data

for j in range(len(ck_10pct)): #For loop that sets all coefficients with an index higher than the first 10% of data to zero
  if j > round(0.1*len(ck_10pct)):
    ck_10pct[j] = 0

for k in range(len(ck_5pct)): #For loop that sets all coefficients with an index higher than the first 5% of data to zero
  if k > round(0.05*len(ck_5pct)):
    ck_5pct[k] = 0

z1 = irfft(ck_10pct) #Inverse Fourier transform of the 10% data array
z2 = irfft(ck_5pct) #Inverse Fourier transform of the 5% data array
time = np.arange(0, len(dow), 1) #Time points for the two FT data points to copy the dimensions of the dow.txt datafile

offset = 4

fig, ax = plt.subplots()
ax.plot(dow/1000, color = 'blue', label = 'Original')
ax.plot(time, z1/1000+offset, color = 'green', linestyle ='--', label = 'Smoothed (10%)')
ax.plot(time, z2/1000-offset, color = 'red', linestyle = '--', label = 'Smoothed (5%)')
ax.set_xlabel("Business days")
ax.set_ylabel("USD (in thousands)")
ax.set_title("Daily Closing Value of Dow Jones Industrial Average")
ax.legend()
plt.show()
````

The plots above show the changes in the daily closing value of Dow Jones Industrial Average for each business day from late 2006 until the end of 2010. Comparing the original plot with the ones where we applied DFT and IDFT, we can see that the curves become smoother as we set more and more coefficients to zero. Since we are effectively removing noise, we end up with a more visually 'pleasant' curve, since the purpose of the data is to see the long term changes of the daily closing value of Dow Jones and not to specifically point out the exact value for any specific day.
