**The Ising model**

_Refer to Exercise 10.9 in the Newman text._ The Ising model is a theoretical model of a magnet. The magnetic potential energy due to the interaction of two dipoles is proportional to their dot product, but in the Ising model this simplifies to just the product $s_i s_j$ for spins on sites $i$ and $j$ of the lattice, since the spins are one-dimensional scalars, not vectors. Then the actual energy of interaction is $-Js_i s_j$, where $J$ is a positive interaction constant. The minus sign ensures that the in- teractions are ferromagnetic, meaning the energy is lower when dipoles are lined up. A ferromagnetic interaction implies that the material will magnetize if given the chance.

Normally it is assumed that spins interact only with those that are immediately adjacent to them on the lattice, which gives a total energy for the entire system equal to
$$ E = -J \sum_{\langle ij \rangle} s_i s_j,$$
where the notation $\langle ij \rangle$ indicates a sum over pairs $i,j$ that are adjacent on the lattice. On the square lattice we use in this exercise each spin has four adjacent neighbors with which it interacts, except for the spins at the edges of the lattice, which have either two or three neighbors.

```python
#Initializing packages
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import *
```

PRELIMINARIES
- Here we are simply defining our constants and choosing a seed for reproducibility of results.

```python
#Constants
N = 20
J = 1
T = 1
kB = 1
beta = 1/(kB*T)
steps = 1000000

seed(69) #Nice
```

First, we create a random initial state for our magnet by representing it as a $20 \times 20$ 2D Numpy array with values $+1$ and $-1$ to represent spin-up and spin-down magnetic dipoles. For the purposes of our problem, we want our initial state to have roughly zero magnetization. We use the `random()` function to choose the spin of each lattice site. This initial state is visualized in the plot below. For this plot and any subsequent plots, spins with values $+1$ and $-1$ are symbolized by the colors **yellow** and **blue**, respectively.

```python
#Creating random initial state with around equal +1 and -1 spins
s = np.empty((N,N), int)
for i in range(N):
  for j in range(N):
    if random()<0.5:
      s[i,j] = 1
    else:
      s[i,j] = -1

#blue = -1, yellow = +1
fig, ax = plt.subplots(figsize = (6,6))
ax.imshow(s, cmap = 'cividis')
ax.set_xticks([])
ax.set_yticks([])
ax.set_title("Random initial state with N = 20")
plt.show()
```

In general, the Ising model prescribes the following Hamiltonian to a given state:

$$
H = -\sum_{<ij>}J_{ij}s_is_j - \mu \sum_j h_j s_j
$$

where $J_{ij}$ is the interaction between two adjacent spins, $\mu$ is the magnetic moment, and $h_j$ is the external magnetic field on the lattice site *j*. For our problem, it is assumed that:

1.   there is no external magnetic field acting on our system $(h = 0)$, and
2.   the interaction strength between all neighbours $<i,j>$ is constant $(J_{ij} = J)$.

Thus, the Hamiltonian or the total energy describing our system simplifies to

$$
H = -J\sum_{<ij>}s_is_j
$$

Because of the negative sign of our Hamiltonian, the interaction $J$ specifies what type of magnetic system we are dealing with. If $J > 0$, then the energy of our system is lowered if a given pair of adjacent spins $s_i$ and $s_j$ are aligned. This interaction describes a *ferromagnetic* system, which corresponds to a material that spontaneously magnetizes if given the chance. If $J < 0$, then the energy of our system is lowered when $s_i$ and $s_j$ have opposite signs. This interaction describes an *antiferromagnetic* system, which describes a material that prefers to have little or no magnetization. For our problem, we consider the case when $J=1$, but we will also consider the case $J=-1$ later on.

Since we represented our magnet as a collection of magnetic dipoles, the total magnetization of our system is the sum of each magnetic dipole.

$$
M = \sum_i s_i
$$

We define the following functions to compute the total energy and magnetization of a given lattice state. For the total energy of a system, since the contribution of any particular pair of adjacent spins only appears once in the sum, we can take dot product of a particular row of spins with the row below it and move on to the next row. The same logic follows for a particular column of spins and the column on its right. We repeat this process until the index $N-2$ or the $(N-1)^{th}$ row/column is dotted with the $N^{th}$ row/column.

```python
def energy(state, J):
  total = 0
  for i in range(N-1): #iterate until N-2, so we dot the 2nd last row/column with the last row/column
    row = np.dot(state[i],state[i+1]) #s_{i,j}*s_{i+1,j} for an entire row at a time
    column = np.dot(state[:, i], state[:, i+1]) #s_{i,j}*s_{i,j+1} for an entire column at a time
    total += (row + column)
  return -J*total


def mag(state):
  return np.sum(state)
```

Now, we can do a Markov Chain Monte Carlo (MCMC) simulation on our system to see its evolution over time. We use the Metropolis-Hastings algorithm to obtain our target distribution or state, which is the lowest energy state of the system. To start, our proposal state is the random initial state `s` we have already defined.

A system with temperature $T$ can occupy multiple states at any given moment.
The probability of obtaining a specific state $i$ with energy (obtained using the Hamiltonian) $E_i$ is given by:

$$
P(E_i) = \frac{e^{-\beta E_i}}{Z}
$$

where $\beta = \dfrac{1}{k_B T}$, $k_B$ is the Boltzmann constant, and $Z$ is the total sum of all the possible states.

For the initial state $s_0$, we generate the next state $s_1$ with energy $E_j$ by randomly choosing a lattice site and inverting the spin at that site. Then, we either accept or reject this new state with the acceptance probability $P_n$ given by

$$
P_n =
\begin{cases}
  1  & \text{if } E_j \leq E_i, \\
  e^{-\beta(E_j - E_i)} & \text{if } E_j > E_i.
\end{cases}
$$

Since our target is the ground state of the system, we accept the new state $s_1$ if its configuration results into a lower energy than when we started. If $s_1$ results in a configuration with a higher energy, we can still accept it given that the difference between $E_j$ and $E_i$ is small enough. Thus, the acceptance probability prescribed is inversely proportional to $E_j- E_i$. For our problem, we nondimensionalize $k_B$ and $T$ set both to be equal to $1$ to simplify things. We can further simplify the problem by rephrasing the acceptance probability as follows:

$$
P_n =
\begin{cases}
  1  & \text{if } dE \leq 0, \\
  e^{-dE} & \text{if } dE > 0.
\end{cases}
$$

where $dE = E_j - E_i$.

```python
E1 = energy(s, J) #Energy of initial state

#Lists for plotting
Epoints = []
Mpoints = []
splots = []

#Metropolis-Hastings simulation
for k in range(steps):
  i = randint(0, N)
  j = randint(0, N)
  s[i,j] *= -1 #Flip spin of random site

  E2 = energy(s, J) #Energy of new state
  dE = E2 - E1 #Change in energy
  if dE < 0: #move accepted
    E1 = E2
    M = mag(s)
  else:
    if random() < np.exp(-beta*dE): #move accepted
      E1 = E2
      M = mag(s)
    else: #move rejected
      s[i,j] *= -1

  Epoints.append(E1)
  Mpoints.append(M)
  splots.append(s.copy())
```

The plots of the expected values of the energy and magnetization of our system as it goes through iterations of the algorithm can be seen below. From our initial state, the energy and magnetization both start near zero. Notice that the burn-in period of our MCMC method goes from $t = 0$ to much less than $t = 1 \times 10^5$, with samples past that being almost homogenous apart from random noise. From our plots we can conclude that in the ground state configuration, the energy of our system is `min(Epoints)` $= - 760$, with a magnetization of $-400$. This magnetization value corresponds to all sites of our $20 \times 20$ lattice having a spin of $-1$.


NOTES:
- Since we nondimensionalized our problem, $[E] = [M] = 1$.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (15, 6))
ax1.plot(Epoints, color = 'g')
ax1.grid(True)
ax1.set_xlabel('Time steps')
ax1.set_ylabel('Energy')
ax2.plot(Mpoints, color = 'r')
ax2.grid(True)
ax2.set_xlabel('Time steps')
ax2.set_ylabel('Magnetization')
fig.tight_layout()
fig.suptitle('Total energy and magnetization of a ferromagnetic material', size = 14, weight = 'bold')
fig.subplots_adjust(top=0.93)
plt.show()
```

The time evolution of the spins our lattice configuration can be seen on the plots below. The fact that we are sampling from the target ground state configuration past $t = 2.5 \times 10^4$ is evident here.

```python
points = [0, 2500, 10000, 25000, 40000, 100000]

plt.figure(figsize=(10,8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(splots[points[i]], cmap = 'cividis')
    plt.xticks([])
    plt.yticks([])
    plt.tight_layout(pad = 2.35)
    plt.suptitle("Time evolution of the spins for a ferromagnetic material", size = 14, weight = 'bold')
    plt.subplots_adjust(top=0.93)
    plt.title(r'$t = $' + str(points[i]))
plt.show()
```

Note that there are two possible ground state configurations for ferromagnetic materials. Since energy is minimized when the spins from two adjacent lattice sites are aligned, it is also possible to end up with a ground state with all sites having a spin of $+1$. It just so happened that more spin-up sites were flipped during the random process of swapping spins, which caused the negative magnetization ground state to dominate.

```python
#Repeating the exact same code but with a different seed corresponding to a different initial state and sites flipped
seed(5344)

s1 = np.empty((N,N), int)
for i in range(N):
  for j in range(N):
    if random()<0.5:
      s1[i,j] = 1
    else:
      s1[i,j] = -1

E3 = energy(s1, J)
M1 = mag(s1)
E1points = []
M1points = []
s1plots = []
for k in range(steps):
  i = randint(0, N)
  j = randint(0, N)
  s1[i,j] *= -1

  E4 = energy(s1, J)
  dE = E4 - E3
  if dE < 0: #move accepted
    E3 = E4
    M1 = mag(s1)
  else:
    if random() < np.exp(-beta*dE): #move accepted
      E3 = E4
      M1 = mag(s1)
    else: #move rejected
      s1[i,j] *= -1

  E1points.append(E3)
  M1points.append(M1)
  s1plots.append(s1.copy())
```

Anyhow, we can also consider the case where the interaction between adjacent lattice sites are *antiferromagnetic* or when $J < 0$. Since it is more energetically favored for adjacent magnetic dipoles to have unaligned spins, the total magnetization of the ground state configuration should tend to be around zero.

```python
#Repeating the exact same code again, but with +J -> -J
s2 = np.empty((N,N), int)
for i in range(N):
  for j in range(N):
    if random()<0.5:
      s2[i,j] = 1
    else:
      s2[i,j] = -1

E5 = energy(s2, -J)
M2 = mag(s2)
E2points = []
M2points = []
s2plots = []
for k in range(steps):
  i = randint(0, N)
  j = randint(0, N)
  s2[i,j] *= -1

  E6 = energy(s2, -J)
  dE = E6 - E5
  if dE < 0: #move accepted
    E5 = E6
    M2 = mag(s2)
  else:
    if random() < np.exp(-beta*dE): #move accepted
      E5 = E6
      M2 = mag(s2)
    else: #move rejected
      s2[i,j] *= -1

  E2points.append(E5)
  M2points.append(M2)
  s2plots.append(s2.copy())
```

The plots of the total energy and magnetization of our antiferromagnetic material can be seen below. Just like before, the energy of our ground state is `min(E2points)` $= -760$, but the magnetization is now $0$.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (15, 6))
ax1.plot(E2points, color = 'g')
ax1.grid(True)
ax1.set_xlabel('Time steps')
ax1.set_ylabel('Energy')
ax2.plot(M2points, color = 'r')
ax2.grid(True)
ax2.set_xlabel('Time steps')
ax2.set_ylabel('Magnetization')
ax2.set_ylim([-100,100])
fig.tight_layout()
fig.suptitle('Total energy and magnetization of an antiferromagnetic material', size = 14, weight = 'bold')
fig.subplots_adjust(top=0.93)
plt.show()
```

As expected, the ground state configuration is the state wherein the spins are alternating at each lattice site, forming a "chessboard"-like pattern.
